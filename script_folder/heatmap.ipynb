```python
#import libraries required for visuliasation
import numpy as np
import pandas as pd 
#import matplotlab 
import matplotlib.pyplot as plt
```

```python
#Drops the null values 
data=pd.read_excel('normalized_readcounts2.xlsx')
# drop the null value 
data_BM=data.dropna(how="any")
#eaxmin data set 
data_BM.head()
```

```python
data.shape
```

```python
data_BM.columns
```

```python
# creating data frame for the larvae 
table = pd.pivot_table(data_BM, values=['Larvae_Rep1', 'Larvae_Rep2', 'Larvae_Rep3'], index=['miRNA_names '])
Larva= pd.DataFrame(table.mean(axis=1))  
Larva.head()
```

```python
# creating data frame for the Pupae 
table = pd.pivot_table(data_BM, values=['Pupae_Rep1', 'Pupae_Rep2', 'Pupae_Rep3'], index=['miRNA_names '])
Pupa= pd.DataFrame(table.mean(axis=1))  
Pupa.head()
```

```python
# creating data frame for the Teneral_females
table = pd.pivot_table(data_BM, values=['Teneralfemales_Rep1', 'Teneralfemales_Rep2', 'Teneralfemales_Rep3'], index=['miRNA_names '])
Teneralfemale= pd.DataFrame(table.mean(axis=1))  
Teneralfemale.head()
```

```python
# creating data frame for the Teneral male 
table = pd.pivot_table(data_BM, values=['Teneralmale_Rep1', 'Teneralmale_Rep2', 'Teneralmale_Rep3'], index=['miRNA_names '])
Teneralmale= pd.DataFrame(table.mean(axis=1))  
Teneralmale.head()
```

```python
# creating data frame for the Adult female
table = pd.pivot_table(data_BM, values=['Adultfemale_Rep1','Adultfemale_Rep2', 'Adultfemale_Rep3'], index=['miRNA_names '])
Adultfemale= pd.DataFrame(table.mean(axis=1))  
Adultfemale.head()
```

```python
#creating data frame for the Adult male
table = pd.pivot_table(data_BM, values=['Adultmale_Rep1', 'Adultmale_Rep2', 'Adultmale_Rep3',], index=['miRNA_names '])
Adultmale= pd.DataFrame(table.mean(axis=1))  
Adultmale.head()
```

```python
#creating data frame for the Gravid female 
table = pd.pivot_table(data_BM, values=['Gravidfemale_Rep1', 'Gravidfemale_Rep2', 'Gravidfemale_Rep3',], index=['miRNA_names '])
Gravidfemale= pd.DataFrame(table.mean(axis=1))  
Gravidfemale.head()
```

```python
#creating data frame for the Gravid female 
table = pd.pivot_table(data_BM, values=['Wildtypemales_Rep1','Wildtypemales_Rep2', 'Wildtypemales_Rep3'], index=['miRNA_names '])
Wildtypemale= pd.DataFrame(table.mean(axis=1))  
Wildtypemale.head()
```

```python
import pandas as pd
from functools import reduce
```

```python
# compile the list of dataframes you want to merge
data_frames = [Larva, Pupa, Teneralfemale,Teneralmale,Adultfemale,Adultmale,Gravidfemale,Wildtypemale]
```

```python
#merged the dataframes (lifestages)
Life_stages = reduce(lambda  left,right: pd.merge(left,right,on=['miRNA_names '],
                                            how='outer'), data_frames)
```

```python
Life_stages.columns =['Larva', 'Pupa', 'Teneral_female', 'Teneral_male','Non_teneral_female','Non_teneral_male','Gravid_female','Wild_type_male']
```

```python
#created a life stage for visualization
Life_stages
```

```python
#installing seaborn programme 
# importing required libraries
import seaborn as sns


import numpy as np
import pandas as pd

# importing matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# cmap = sns.light_palette("color-code",as_cmap=True)
plt.figure(figsize=(20,40))
sns.heatmap(Life_stages, cmap="coolwarm", annot=False, vmin=0, vmax=20000)
```

```python
# Hierarchically-clustered Heatmap in Python with Seaborn Clustermap
sns.clustermap(Life_stages, cmap="coolwarm", annot=False, vmin=0, vmax=10000,figsize=(15, 20))
```
